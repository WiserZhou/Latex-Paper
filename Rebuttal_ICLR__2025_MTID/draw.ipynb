```python
import pandas as pd
import matplotlib.pyplot as plt

# 定义科研绘图风格参数
plot_params = {
    'figure.figsize': (10, 6),
    'font.family': 'Times New Roman',
    'font.size': 12,
    'axes.labelsize': 14,
    'axes.titlesize': 16,
    'axes.linewidth': 1.5,
    'lines.linewidth': 2,
    'lines.markersize': 8,
    'legend.fontsize': 12,
    'xtick.labelsize': 12,
    'ytick.labelsize': 12,
    'grid.linestyle': '--',
    'grid.alpha': 0.3
}

# 应用绘图参数
plt.rcParams.update(plot_params)

# 读取5个CSV文件
files = ['a0.csv', 'a1.csv', 'a3.csv', 'a4.csv', 'a5.csv']
dataframes = []

for file in files:
    try:
        df = pd.read_csv(file)
        dataframes.append(df)
    except FileNotFoundError:
        print(f"警告: 找不到文件 {file}")

# 设置图表样式
fig, ax = plt.subplots()
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']  # 专业配色方案

# 绘制每个文件的折线图
# 绘制每个文件的折线图
for i, df in enumerate(dataframes):
    ax.plot(df['Step'], 
            df.iloc[:, 1],
            label=f'T={i}',
            color=colors[i],
            linewidth=2)  # 移除了 marker 相关的参数

# 设置图表属性
# 设置图表属性
# ax.set_title('Action Validation Accuracy Comparison', pad=15)
# ax.set_xlabel('Training Steps')
ax.set_ylabel('Accuracy (%)', fontsize=25)
ax.grid(True, linestyle='--', alpha=0.3)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(direction='out', length=6, width=1.5)
# 修改 legend 位置，移除 bbox_to_anchor，使用更合适的位置参数
ax.legend(frameon=False, loc='lower right')

# 调整布局
plt.tight_layout()
plt.show()
```
