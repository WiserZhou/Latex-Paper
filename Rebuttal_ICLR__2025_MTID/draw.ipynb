```python
import pandas as pd
import matplotlib.pyplot as plt

# 
plot_params = {
    'figure.figsize': (10, 6),
    'font.family': 'Times New Roman',
    'font.size': 12,
    'axes.labelsize': 14,
    'axes.titlesize': 16,
    'axes.linewidth': 1.5,
    'lines.linewidth': 2,
    'lines.markersize': 8,
    'legend.fontsize': 12,
    'xtick.labelsize': 12,
    'ytick.labelsize': 12,
    'grid.linestyle': '--',
    'grid.alpha': 0.3
}

# 
plt.rcParams.update(plot_params)

# 5CSV
files = ['a0.csv', 'a1.csv', 'a3.csv', 'a4.csv', 'a5.csv']
dataframes = []

for file in files:
    try:
        df = pd.read_csv(file)
        dataframes.append(df)
    except FileNotFoundError:
        print(f":  {file}")

# 
fig, ax = plt.subplots()
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']  # 

# 
# 
for i, df in enumerate(dataframes):
    ax.plot(df['Step'], 
            df.iloc[:, 1],
            label=f'T={i}',
            color=colors[i],
            linewidth=2)  #  marker 

# 
# 
# ax.set_title('Action Validation Accuracy Comparison', pad=15)
# ax.set_xlabel('Training Steps')
ax.set_ylabel('Accuracy (%)', fontsize=25)
ax.grid(True, linestyle='--', alpha=0.3)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(direction='out', length=6, width=1.5)
#  legend  bbox_to_anchor
ax.legend(frameon=False, loc='lower right')

# 
plt.tight_layout()
plt.show()
```
